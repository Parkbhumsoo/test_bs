## 1. 부부의세계 article

```
import requests
from bs4 import BeautifulSoup
```

```
url = "https://news.joins.com/article/23755886/"  # 부부의 세계 기사
res= requests.get(url)
soup=BeautifulSoup(res.text, 'html.parser')
```

```
top = soup.find('div',id="article_body")
a=top.text
```

```
print(a.strip())   #기사 내용
```

```
bot = soup.find('h1',{'class':'headline mg'})
bot2=bot.text
```

```
print(bot2) # 기사 제목
```
