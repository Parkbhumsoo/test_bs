```
import pandas as pd
import numpy as np


df=pd.DataFrame(np.random.randn(3,3),index=['a','c','e'],columns=['one','two','three'])
```

```
df
```

```
df=df.reindex(['a','b','c'])   # index를 바꿔주고 없던 값은 NAN으로 됌
df
```

```
df = df.fillna(1)   ### NAN이란 표시를 채워줌
df
```

```
df4=pd.DataFrame(np.random.randn(5,3),index=['a','c','e','f','h'],columns=['one','two','three'])
df4
```

```
df4=df4.reindex(['a','b','c','d','e','f','g','h'])
df4
```
