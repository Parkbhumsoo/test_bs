```
#{'num':'1', 'name':'김철수','kor':90, 'eng':80, 'math':85, 'total':0, 'avg':0.0}
a= np.array([[1, 90, 80, 85, 0, 0.0, 0],[2, 90, 85, 90, 0, 0.0, 0],[3, 80, 80, 80, 0, 0.0, 0]],dtype=np.int64)
```

```
a,type(a)
```

```
a[:,4]=np.sum(a[:,1:4],axis=0)     #total값 생성
```

```
a
```

```
a[:,5]=np.array(a[:,4]/3)            #평균값 생성
a
```

```
b=np.sum(a[:,1]/3,dtype=np.int64)
print("국어점수의 평균은",b,"점입니다.")
c=np.sum(a[:,2]/3,dtype=np.int64)
print("영어점수의 평균은",c,"점입니다.")
d=np.sum(a[:,3]/3,dtype=np.int64)
print("수학점수의 평균은",d,"점입니다.")
```
