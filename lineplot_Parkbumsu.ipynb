```
import numpy as np
from matplotlib import pyplot as plt

student_score=[
               [1,90,80,90,0,0.0,0,0],
               [2,70,50,70,0,0.0,0,0],
               [3,50,70,50,0,0.0,0,0],
               [4,30,50,30,0,0.0,0,0],
               [5,10,30,10,0,0.0,0,0]
]
```

```
st=np.array(student_score)
st[1,1:4]
```

```
x1=['Kor','Eng','Math']
y1=st[0,1:4]
x2=['Kor','Eng','Math']
y2=st[1,1:4]
x3=['Kor','Eng','Math']
y3=st[2,1:4]
x4=['Kor','Eng','Math']
y4=st[3,1:4]
x5=['Kor','Eng','Math']
y5=st[4,1:4]
plt.bar(x1, y1, label="student1", color='b')
plt.bar(x2, y2, label="student2", color='g')
plt.bar(x3, y3, label="student3", color='r')
plt.bar(x4, y4, label="student4", color='y')
plt.bar(x5, y5, label="student5", color='c')
plt.xlabel("students")
plt.ylabel("score")
plt.title("student_score")
plt.legend()
plt.show()
```

```
n = 5 + np.random.randn(1000)
m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()
```

```
plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()
```

```
plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```
