```
y1=si.iloc[0,1:]
y2=si.iloc[1,1:]
y3=si.iloc[2,1:]
x1=[2018,2019,2020,2021,2022,2023,2024]
plt.plot(x1,y1,'bo-',label = 'sum sales')
plt.plot(x1,y2,'ro-',label = 'direct sales')
plt.plot(x1,y3,'ko-',label = 'indirect sales')
plt.xlabel("year")
plt.ylabel("Sales")
plt.title("dater industry market outlook")
plt.legend()
plt.show()
```
