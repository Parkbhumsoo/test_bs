```
ax = income2.plot(kind='bar', title='a personal income',color='0', figsize=(12, 4), legend=True, fontsize=12)
ax.set_xlabel('city',fontsize=12)
ax.set_ylabel('a personal income',fontsize=12)
```

```
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (16,6)
plt.rc('font', family ='Malgun Gothic')
x=income['시도별']
y=income['2018']
y2=income['2017']
y3=income['2016']
y4=income['2015']
y5=income['2014']
plt.plot(x,y,'ro--',label='2018')
plt.plot(x,y2,'go--',label='2017')
plt.plot(x,y3,'yo--',label='2016')
plt.plot(x,y4,'ko--',label='2015')
plt.plot(x,y5,'bo--',label='2014')
plt.show()
```
