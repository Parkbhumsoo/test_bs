```
for i in range(len(wine.feature_names)):
  print(wine.feature_names[i])
  print("평균:",round(np.mean(wine.data[:,i]),1))
  print("중간값:",round(np.median(wine.data[:,i]),1))
  print("표준편차:",round(np.std(wine.data[:,i]),1))
  print("최대값:",np.max(wine.data[:,i]))
  print("최소값:",np.min(wine.data[:,i]))
  print("\n")
```

평균과 중간값의 차이가 크지 않고 

중간값과 평균의 차가 표준편차를 넘지 않기 때문에 대표값을 평균으로 정했습니다.

```
for i in range(len(wine.feature_names)):  
  print("대표값은:",wine.feature_names[i],"의 평균인",round(np.mean(wine.data[:,i]),1),"입니다.")
```
