#IRIS (붓꽃) 데이터

iris의 datasets = ['data','target','target_names','DESCR','feature_names','filename'] 의 총6열로 구성.

샘플(data): 150개

타겟 값(target_names): setosa,versicolor,verginica

속성:꽃받침 길이(sepal length),꽃받침 폭(sepal width),

꽃잎 길이(petal length),꽃잎 폭(petal width)

```
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
iris=datasets.load_iris()
```

```
iris['target']
iris.keys()
```

```
iris['target_names']
iris['feature_names']
```

```
sts=iris.data[iris.target==0]
vsc=iris.data[iris.target==1]
vgn=iris.data[iris.target==2]
```

```
for i in range(len(iris.feature_names)):
    print("setosa",iris.feature_names[i])
    print("평균:",round(np.mean(sts[:,i]),1))
    print("중간값:",round(np.median(sts[:,i]),1))
    print("표준편차:",round(np.std(sts[:,i]),1))
    print("최대값:",np.max(sts[:,i]))
    print("최소값:",np.min(sts[:,i]))
    print("\n")

for i in range(len(iris.feature_names)):
    print("versicolor",iris.feature_names[i])
    print("평균:",round(np.mean(vsc[:,i]),1))
    print("중간값:",round(np.median(vsc[:,i]),1))
    print("표준편차:",round(np.std(vsc[:,i]),1))
    print("최대값:",np.max(vsc[:,i]))
    print("최소값:",np.min(vsc[:,i]))
    print("\n")

for i in range(len(iris.feature_names)):
    print("verginica",iris.feature_names[i])
    print("평균:",round(np.mean(vgn[:,i]),1))
    print("중간값:",round(np.median(vgn[:,i]),1))
    print("표준편차:",round(np.std(vgn[:,i]),1))
    print("최대값:",np.max(vgn[:,i]))
    print("최소값:",np.min(vgn[:,i]))
    print("\n")
```

평균값과 중간값을 비교했을때 거의 일치하고 표준편차값도 낮아 대표값은 평균으로 정했습니다.

```
for i in range(len(iris.feature_names)):
  print(iris.feature_names[i])
  print("iris의 대표값은 평균점수인:",round(np.mean(iris.data[:,i]),1),"입니다.")
```

```
sepals_length = iris.data[:,0]
```

```
x1=sepals_length[iris.target==0]
x2=sepals_length[iris.target==1]
x3=sepals_length[iris.target==2]
```

```
sepals_width=iris.data[:,1]
sepals_width
```

```
y1=sepals_width[iris.target==0]
y2=sepals_width[iris.target==1]
y3=sepals_width[iris.target==2]
```

```
y1
```

꽃받침의 길이 너비

```
plt.scatter(x1,y1,marker='o',label='setosa')
plt.scatter(x2,y2,marker='v',color='red',label='versicolor')
plt.scatter(x3,y3,marker='^',color='m',label='virginica')
plt.title('sepals lenth and width')
plt.xlabel("length")
plt.ylabel("width")
plt.legend()
plt.show()
```

```
petals_length = iris.data[:,2]
```

```
x4=petals_length[iris.target==0]
x5=petals_length[iris.target==1]
x6=petals_length[iris.target==2]
```

```
petals_width=iris.data[:,3]
```

```
y4=petals_width[iris.target==0]
y5=petals_width[iris.target==1]
y6=petals_width[iris.target==2]
```

```
plt.scatter(x4,y4,marker='o',label='setosa')
plt.scatter(x5,y5,marker='v',color='blue',label='versicolor')
plt.scatter(x6,y6,marker='^',color='yellow',label='virginica')
plt.title('petals lenth and width',fontsize=20)
plt.xlabel("length",fontsize=14)
plt.ylabel("width",fontsize=14)
plt.legend()
plt.show()
```

```
a1=np.mean(x1)
a2=np.mean(x2)
a3=np.mean(x3)
b1=np.mean(y1)
b2=np.mean(y2)
b3=np.mean(y3)
```

```
plt.scatter(a1,b1,marker='o',label='setosa')
plt.scatter(a2,b2,marker='v',color='red',label='versicolor')
plt.scatter(a3,b3,marker='^',color='m',label='virginica')
plt.title('sepals lenth and width mean')
plt.xlabel("length")
plt.ylabel("width")
plt.legend()
plt.show()
```
