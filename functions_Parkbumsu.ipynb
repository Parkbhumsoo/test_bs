```
def sumf(a,b):                    #덧셈
   result=int(a)+int(b)
   print("Inside the function :",result)
   return result;
```

```
first,second=input("input two Number?").split(',')
total = sumf(first,second)
print("Outside the function:",total)
```

```
def summ(c,d):                   #뺄셈
   result=int(c)-int(d)
   print("Inside the function :",result)
   return result;
```

```
first,second=input("input two Number?").split(',')
total = summ(first,second)
print("Outside the function:",total)
```

```
def summu(e,f):                 #곱셈
   result=int(e)*int(f)
   print("Inside the function :",result)
   return result;
```

```
first,second=input("input two Number?").split(',')
total = summu(first,second)
print("Outside the function:",total)
```

```
def sumd(g,h):                 #나눗셈
   result=int(g)/int(h)
   print("Inside the function :",result)
   return result;
```

```
first,second=input("input two Number?").split(',')
total = sumd(first,second)
print("Outside the function:",total)
```
